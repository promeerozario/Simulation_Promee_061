```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
arrivals = np.array([
    8,12,10,15,9,13,11,14,7,16,
    10,12,9,15,11,13,14,8,12,10
])

print(arrivals)
```

```python
mean_arrival = np.mean(arrivals)

population_variance = np.var(arrivals)

sample_variance = np.var(arrivals, ddof=1)

population_std = np.std(arrivals)

sample_std = np.std(arrivals, ddof=1)

print("Mean =", mean_arrival)
print("Population Variance =", population_variance)
print("Sample Variance =", sample_variance)
print("Population Std =", population_std)
print("Sample Std =", sample_std)
```

```python
np.random.seed(42)

uniform_data = np.random.uniform(
    low=np.min(arrivals),
    high=np.max(arrivals),
    size=30
)

print(uniform_data)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(arrivals, bins=6)
plt.title("Real Arrivals")

plt.subplot(1,2,2)
plt.hist(uniform_data, bins=6)
plt.title("Simulated Uniform Data")

plt.tight_layout()
plt.show()
```

```python
mc_samples = np.random.uniform(
    low=np.min(arrivals),
    high=np.max(arrivals),
    size=100000
)

probability = np.mean(mc_samples > 15)

print("Probability(arrivals > 15) =", probability)
```

```python
last_3_digits = 61

hyp_mean = (last_3_digits % 10) + 10

t_stat, p_value = stats.ttest_1samp(
    arrivals,
    hyp_mean
)

print("Hypothesized Mean =", hyp_mean)
print("t-statistic =", t_stat)
print("p-value =", p_value)

alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
plt.figure(figsize=(6,5))

plt.axhline(mean_arrival, label='Sample Mean')

plt.axhline(
    hyp_mean,
    linestyle='--',
    label='Hypothesized Mean'
)

plt.title("Mean Comparison")

plt.legend()

plt.show()
```
